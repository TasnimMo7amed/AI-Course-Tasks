```python
import numpy as np
import pandas as pd
import scipy.stats as spss
import plotly.express as px
import seaborn as sns
```

#  Types of distributions in statistics

##    Continuous Distributions


****

### **Uniform distribution**

As the name suggests, in uniform distribution the probability of all outcomes is same. The shape of this distribution is a rectange.


```python
rv_array = spss.uniform.rvs(size=10000, loc = 10, scale=20)
ax = sns.distplot(rv_array, kde= True, color='black',hist= False)
ax.set(xlabel='value_of_random_variable', ylabel='Count')
```

### Normal Distribution, also known as Gaussian distribution:

1-The normal distribution is the most important probability distribution in statistics because it fits many natural phenomena.

2-Normal distribution is a limiting case of Poisson distribution with the parameter lambda tends to infinity.

3-since poisson distribution is a form of binomial distribution, normal distribution is also a form of binomial distribution. 

4-This distribution has a bell-shaped density curve described by its mean and standard deviation. 

5-The mean represents the location and the sd represents the spread of the distribution. 

6-The curve represents that the data near the mean occurrs more frequently than the data far from the mean.

#### Standard Normal Distribution

Is a special case of the normal distribution where mean = 0 and sd = 1

```python
n = spss.norm.rvs(size=10000,loc=0,scale=1)
bx = sns.distplot(n,  kde= True, color='green',hist= False)
bx.set(xlabel='Normal Distribution', ylabel='Frequency')
```

### Non-Normal Distribution

#### Gamma distribution 

1-Gamma distribution is a two-parameter family of continuous probability distributions

2-<font color="red">**Exponential**</font>, <font color="Blue">**chi-squared**</font>, <font color="green">**Erlang** </font> distributions are special cases of the gamma distribution

```python
Gamma  = spss.gamma.rvs(a=5, size=10000) # size specifies number of random variates, a is the shape parameter
gx=sns.distplot(Gamma , kde=True,color="red", hist= False)
gx.set(xlabel='Gamma Distribution', ylabel='Density')
```

##### Erlang distribution

Special case of Gamma distribution when **a is an integer**.

#### Exponential distribution

1-Special case of Gamma distribution with **a=1** . 

2-Exponential distribution describes the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate.

```python
Exponential = spss.expon.rvs(scale=1,loc=0,size=1000) # size specifies number of random variates, loc corresponds to mean, scale corresponds to standard deviation
EX=sns.distplot(Exponential, kde=True, color="SkyBlue",hist= False)
EX.set(xlabel='Exponential Distribution', ylabel='Density')
```

### Beta Distribution


```python
Beta = np.random.beta(a=4, b=15, size=1000)
Bex=sns.distplot(Beta , kde=True,color="Brown", hist= False)
Bex.set(xlabel='Beta Distribution', ylabel='Density')
```

## Discrete Distributions

### Binomial Distribution

1-Distribution where only two outcomes are possible, such as success or failure, gain or loss, win or lose. 

2-The probability of success and failure is same for all the trials. 

3-The outcomes need not be equally likely, and each trial is independent of each other.

4-The probability of observing k events in an interval is given by the equation: 

  <font color="red">f(k;n,p) = nCk * (p^k) * ((1-p)^(n-k))</font>
    
  **Where:** 
  
      nCk = (n)! / ((k)! * (n-k)!) 
      n=total number of trials 
      p=probability of success in each trial

```python
Binomial = spss.binom.rvs(n=10,p=0.6,size=1000) # n = number of trials, p = probability of success, size = number of times to repeat the trials
BX=sns.distplot(Binomial, kde=False, color="RosyBrown")
```

### Poisson Distribution

1-Poisson random variable is typically used to model the number of times an event happened in a time interval.

2-Poisson distribution is described in terms of the rate (μ) at which the events happen. 

3-The average number of events in an interval is designated λ (lambda). 

4-Lambda is the event rate, also called the rate parameter. 

5-The probability of observing k events in an interval is given by the equation: 

 <font color="red">P(k events in interval) = e^(-lambda) * (lambda^k / k!)</font>
    
6-Poisson distribution is a limiting case of a binomial distribution under the following conditions:

   **A-** The number of trials is indefinitely large or n tends to infinity.
   
   **B-** The probability of success for each trial is same and indefinitely small or p tends to zero.
   
   **C-** np = lambda, is finite.

```python
PD = spss.poisson.rvs(mu=5, size=10000) # size specifies number of random variates, loc corresponds to mean, scale corresponds to standard deviation
sns.distplot(PD, kde=False, color="SeaGreen")
```

### Bernoulli distribution

1-It has only two possible outcomes, 1 (success) and 0 (failure), and a single trial, for example, a coin toss.

2-The random variable **X** which has a Bernoulli distribution can take value 1 with the probability of success, p, and the value 0 with the probability of failure, q or 1-p.

3-The probabilities of success and failure need not be equally likely. Probability mass function of Bernoulli distribution:

<font color="red">f(k;p) = (p^k) * ((1-p)^(1-k))</font>

4-Bernoulli distribution is a special case of the binomial distribution where a single trial is conducted (n=1).

```python
Bernoulli = spss.bernoulli.rvs(size=10000,p=0.6) # p = probability of success, size = number of times to repeat the trial
Berx=sns.distplot(Bernoulli, kde=False, color="MediumTurquoise")
```
