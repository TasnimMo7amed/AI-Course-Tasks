### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo= pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
chipo
```

### Step 4. How many products cost more than $10.00?

```python
#Convert Iteam price from string to float
chipo['item_price_regex']=chipo['item_price'].str.replace('[\$\,]','',regex=True)
chipo['item_price_regex']=pd.to_numeric(chipo['item_price_regex'])
chipo['item_price_regex']
```

```python
np.count_nonzero(chipo.item_price_regex>10.00)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_name','item_price']]
```

```python
chipo1 = chipo.groupby(by ="item_name")[["item_price"]].min().reset_index()
chipo1
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by = "item_name",ascending=True).reset_index(drop=True)
```

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.groupby(by="item_price").max().sort_values('quantity',ascending=False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
VS= np.count_nonzero(chipo.item_name == "Veggie Salad Bowl")
print(VS,"times a Veggie Salad Bowl was ordered")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
Cann = np.count_nonzero((chipo.item_name == "Canned Soda") & (chipo.quantity > 1))
print(Cann,"times people orderd more than one Canned Soda")
```
