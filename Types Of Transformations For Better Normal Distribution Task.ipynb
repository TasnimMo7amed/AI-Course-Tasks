```python
import numpy as np
import pandas as pd
import scipy.stats as spss
import plotly.express as px
import seaborn as sns
```

# Types Of Transformations For Better Normal Distribution

    Numerical variables may have high skewed and non-normal distribution caused by outliers,highly exponential distribution.Therefore we go for data transformation. Here are the types of transformations for better normal distribution that will be applied on an Exponential distribution as an Original distribution to show the effect of each type on it:


    1-Log Transformation
    2-Square-Root Transformation
    3-Reciprocal Transformation
    4-Box-Cox Transformation
    5-Yeo-Johnson Transformation (Bonus)

## **The Original Distribution**

```python
#make this example reproducible
np.random.seed(0)

#create Exponential distributed random variable with 10000 values

Exponential = spss.expon.rvs(scale=10,loc=0,size=1000) # size specifies number of random variates, loc corresponds to mean, scale corresponds to standard deviation

EX=sns.histplot(Exponential,color="gray")

EX.set(xlabel='Exponential Distribution', ylabel='Density')
```

## Log Transformation :

1-In Log transformation each variable of x will be replaced by log(x) with base 10, base 2, or natural log.

```python
#create log-transformed data
log_target = np.log1p(Exponential)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms

axs[0].hist(Exponential, edgecolor='white', color="SkyBlue")

axs[1].hist(log_target, edgecolor='white', color="SkyBlue")

#add title to each histogram

axs[0].set_title('Original Data')
axs[1].set_title('Log-Transformed Data')
```

## Square-Root Transformation :

1- This transformation will give a moderate effect on distribution. 

2- The main advantage of square root transformation is, it can be applied to zero values.

```python
#create Square-Root-transformed data
sqrt_target=(Exponential)**(1/2)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)


#create histograms
axs[0].hist(Exponential, edgecolor='white',color="lightGreen")
axs[1].hist(sqrt_target, edgecolor='white',color="lightGreen")

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Square Root Transformed Data')
```

## Cube Root Transformation

```python
#create log-transformed data
Cube_target = np.cbrt(Exponential)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(Exponential, edgecolor='white',color="Gold")
axs[1].hist(Cube_target, edgecolor='white',color="Gold")

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Cube Root Transformed Data')
```

## Reciprocal Transformation :

1-In this transformation, x will replace by the inverse of x (1/x).

2- The reciprocal transformation will give little effect on the shape of the distribution.

3-This transformation can be only used for non-zero values.

```python
#create Reciprocal-transformed data
Exponentiall = spss.expon.rvs(scale=1,loc=1,size=10000)
reciprocal_target= 1/(Exponentiall)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(Exponentiall, edgecolor='white',color="LightSalmon")
axs[1].hist(reciprocal_target, edgecolor='white',color="LightSalmon")

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Reciprocal Transformed Data')
```

## Box-Cox Transformation:

1-A Box Cox transformation is a transformation of non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data is not normal, applying a Box-Cox means that you are able to run a broader number of tests. 

2-At the core of the Box Cox transformation is an exponent, lambda (λ), which varies from -5 to 5. All values of λ are considered and the optimal value for your data is selected; The “optimal value” is the one which results in the best approximation of a normal distribution curve.
```python
#create box-cox-transformed data
from scipy.stats import boxcox
bcx_target, lam = boxcox(Exponential)
#lam is the best lambda for the distribution

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)


#create histograms
axs[0].hist(Exponential, edgecolor='white',color="DarkSlateGray")
axs[1].hist(bcx_target, edgecolor='white',color="DarkSlateGray")

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Box-Cox Transformed Data')
```

## Yeo-Johnson Transformation:

1-This is one of the older transformation technique which is very similar to Box-cox transformation but does not require the values to be strictly positive.

2-This transformation is also having the ability to make the distribution more symmetric.

```python
#create Yeo-Johnson-transformed data
from scipy.stats import yeojohnson
yf_target, lam = yeojohnson(Exponential)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)


#create histograms
axs[0].hist(Exponential, edgecolor='white',color="PaleVioletRed")
axs[1].hist(yf_target, edgecolor='white',color="PaleVioletRed")

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Yeo-Johnson Transformed Data')
```
