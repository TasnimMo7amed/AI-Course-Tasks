```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
df
```

```python
# Default sort
df1 = df.sort_values('occupation')
df1
```

```python
# Sort by Descending
df2 = df.sort_values('occupation', ascending=False)
df2
```

```python
# Sory by multiple columns
df3 = df.sort_values(by=['occupation','age'])
df3
```

```python
# Sort and ignore index
df4 = df.sort_values(by='occupation', ignore_index=True)
df4
```

```python
# Sory by putting NaN at first
df5= df.sort_values(by=['occupation','age'], na_position='first')
df5
```

```python
# Sort by function
df6 = df.sort_values(by='occupation', key=lambda col: col.str.lower())
df6
```

```python
# Sort by heap algorithm
df7 = df.sort_values(by='occupation', kind='heap')
df7
```
