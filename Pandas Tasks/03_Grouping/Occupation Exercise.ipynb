### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by ="occupation")[["age"]].mean().reset_index()
```

```python
users.groupby(by="occupation").age.mean().reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#solution1:
#step1
users['Male'] = users.gender.apply(lambda x: True if x == 'M' else False)
users
```

```python
#step2
a=users.groupby('occupation').Male.sum()
a
```

```python
#step3
b=(users.groupby(by="occupation").gender.count())
b
```

```python
#step4
r=(a/b).round(3)
r.sort_values(ascending=False).reset_index()
```

```python
#solution2
users['male']= np.where(users.gender == 'M', True, False)
users.male
```

```python
(users.groupby('occupation').male.sum()/users.groupby('occupation').gender.count()).round(3).sort_values(ascending = False).reset_index()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
#Solution1
Max_age=(users.groupby(by="occupation").age.max())
Min_age=(users.groupby(by="occupation").age.min())
print(Min_age,Max_age)
```

```python
#Solution2
users.groupby(by='occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).age.mean().round(3).reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
Genders = (users.groupby(by="occupation")["gender"].value_counts(normalize=True)*100).round(3)
Genderper = pd.DataFrame(Genders)
Genderper
```
