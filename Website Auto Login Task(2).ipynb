```python
from selenium import webdriver
from getpass import getpass
```

```python
username = input("Enter your username: ")
password = getpass("Enter your password: ")

driver = webdriver.Chrome("D:\AI-Course\Py.Prac\\chromedriver.exe")
driver.get("https://www.github.com/login")

User_Name=driver.find_element_by_xpath('//*[@id="login_feild"]')
User_Name.send_keys(username)

User_Password = driver.find_element_by_xpath('//*[@id="PassWord"]')
User_Password.send_keys(password)

button = driver.find_element_by_id("u_0_3")
button.click()
```
